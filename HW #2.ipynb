#2. generate n by n matrix, n=100, then assign the first n prime numbers in the first row

```python
n=1000
tree = zeros((n, n))

def isPrime(x):
    if x<=3:
        if x==1:
            return False;
        return True;
    
    if not mod(x, 2) or not mod(x, 3):
        return False;
    
    for k in range(5, int(x**0.5)+1, 6):
        if not mod(x, k) or not mod(x, k+2):
            return False;
    
    return True;

# prof
# N= 10
# tree = zeros((N, N))
# index = -1
# for k in range(2, 1000):
#     if isPrime(k):
#         index = index+1
#         tree[0, index]=k
#     if index == N-1:
#         break;
    

# 내가 쓴 버전,,
t=[]
def counting(m):
    n = 0; k=1;
    while(n<m):
        k+=1;
        if(isPrime(k)):
            n += 1;
            t.append(k)   

counting(n)

for i in range(n):
    tree[0, i] = t[i]


matshow(tree)
show()
```

```python
for j in range(1, n):
    for i in range(j, n):
        tree[j, i] = tree[j-1, i] - tree[j-1, i-1];

print(tree)
matshow(tree)
show()
```

```python
tree = abs(tree) #절대값으로 다 바꾸기
print(tree)
matshow(tree[1:n, :]) #첫줄 빼고
show()
```

```python
for i in range(1, n):
    #i층에서 i~9
    tree[i,i:n] = abs(tree[i-1, i:n] - tree[i-1, i-1:n-1])

print(tree)
matshow(tree[1:n, :])
show()
```

```python
for i in range(n):
    for j in range(n):
        if(tree[i, j]>0):
            tree[i, j]=1

matshow(tree[1:n, :])
show()
```

#3. population dynamics - 짚신벌레는 분열법 생식 - exponential growth

```python
data = loadtxt('pop.txt')
t=data[:,0]
p=data[:,1]
n = len(t)
d = zeros(n)

for i in range(1, n): #짚신벌레 개체수가 늘어나는 차이값의 양상
    d[i-1] = p[i]-p[i-1]
    plot(p[i-1], d[i-1], 'or')

show()
```
