```python
from pylab import *
x=linspace(0, 1, 100)
y=sin(pi*x)
plot(x, y, 'ro') #o- : line and circle with bisic color blue // ro- red
xlabel('distance')
ylabel('height')
title('sample')
show()
```

```python
r=1
t = linspace(0, 2*pi, 100)
x=r*cos(t)
y = r*sin(t)
plot(x, y, 'r-')
show()
```

```python
x = linspace(-1, 1, 50)
y = abs(x)
plot(x, y, 'r-')
show()
```

```python
%pylab
x = linspace(0, 1, 100); y = 4*x*(1-x);
for i in range(100):
    clf()
    plot(x[i], y[i], 'o')
    plot(x, y, 'r-')
    axis([0, 1, 0, 1.2])
    pause(0.01)
```
